#### grouping and aggregation

```python
#grouping and aggregation is useful in performing the some calculations based on some common future
#we use groupby() method to do these
monthly_groups=covid_df.groupby('month')
```

```python
type(monthly_groups)
```

```python
monthly_groups[['new_cases','new_tests','new_deaths']].sum()
#.sum() is an aggregation
```

###### average number of cases in a weekday

```python
covid_mean_weekday_df=covid_df.groupby('weekday')[['new_cases','new_deaths','new_tests']].mean()
```

```python
covid_mean_weekday_df
```

```python
#cumsum() method to get total cases upto a particular date
covid_df['total_cases']=covid_df.new_cases.cumsum()
covid_df['total_tests']=covid_df.new_tests.cumsum()
covid_df['total_deaths']=covid_df.new_deaths.cumsum()
```

```python
covid_df
```

#### merging data from mulyiple sources

```python
locations_df=pd.read_csv('locations.csv')
```

```python
locations_df
```

```python
locations_df[locations_df.location=='Italy']
```

```python
covid_df['location']='Italy'
```

```python
covid_df
```

```python
#merge() method for merging the both dataframes so that each should have a common columns so that it merges according to columns parameter
merged_df=covid_df.merge(locations_df,on='location')
```

```python
merged_df
```

```python
#calculating cases,deaths,tests per million
merged_df['cases_per_million']=merged_df.new_cases*1e6/merged_df.population
merged_df['tests_per_million']=merged_df.new_tests*1e6/merged_df.population
merged_df['deaths_per_million']=merged_df.new_deaths*1e6/merged_df.population
```

```python
merged_df
```

```python
#savingthe file which we analysed lastly
#saving the wanted columns in a new data frame
result_df=merged_df[['new_tests','total_tests','new_cases','total_cases','new_deaths','total_deaths','tests_per_million','cases_per_million','deaths_per_million']]
```

```python
result_df
```

```python
#saving the file
#df.to_csv('file_name',index=None)
result_df.to_csv('result.csv',index=None)
#index=None/False doesnt include the indexex in the csv
```
