```python
#for s9 and s10 sensors comparing the RH and Temp by two sensors


plt.figure(figsize=(12,12))
plt.plot(sdict['S9']['AHT_TEMP'],sdict['S9']['DHT_TEMP'],'b.')
plt.title('AHT_TEMP vs DHT_TEMP')
plt.xlabel('AHT_TEMP')
plt.ylabel('DHT_TEMP')
plt.show()


plt.figure(figsize=(12,12))
plt.plot(sdict['S9']['AHT_RH'],sdict['S9']['DHT_RH'],'b.')
plt.title('AHT_RH vs DHT_RH')
plt.xlabel('AHT_RH')
plt.ylabel('DHT_RH')
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.plot(sdict['S10']['AHT_TEMP'],sdict['S10']['DHT_TEMP'],'b.')
plt.title('AHT_TEMP vs DHT_TEMP')
plt.xlabel('AHT_TEMP')
plt.ylabel('DHT_TEMP')
plt.show()


plt.figure(figsize=(12,12))
plt.plot(sdict['S10']['AHT_RH'],sdict['S10']['DHT_RH'],'b.')
plt.title('AHT_RH vs DHT_RH')
plt.xlabel('AHT_RH')
plt.ylabel('DHT_RH')
plt.show()
```

```python
for key in sdict.keys():
    aht_rh=(sdict[key][["AHT_RH","DHT_RH","AHT_TEMP","DHT_TEMP"]].isnull().sum()/sdict[key].shape[0]).values[0]
    dht_rh=(sdict[key][["AHT_RH","DHT_RH","AHT_TEMP","DHT_TEMP"]].isnull().sum()/sdict[key].shape[0]).values[1]
    aht_temp=(sdict[key][["AHT_RH","DHT_RH","AHT_TEMP","DHT_TEMP"]].isnull().sum()/sdict[key].shape[0]).values[2]
    dht_temp=(sdict[key][["AHT_RH","DHT_RH","AHT_TEMP","DHT_TEMP"]].isnull().sum()/sdict[key].shape[0]).values[3]
    if (aht_rh==1 and dht_rh==0):
        sdict[key]['RH']=sdict[key]["DHT_RH"]
    elif (aht_rh==0 and dht_rh==1):
        sdict[key]['RH']=sdict[key]["AHT_RH"]
    else:
        sdict[key]['RH']=(sdict[key]["AHT_RH"]+sdict[key]['DHT_RH'])/2
    if (aht_temp==1 and dht_temp==0):
        sdict[key]['TEMP']=sdict[key]["DHT_TEMP"]
    elif (aht_temp==0 and dht_temp==1):
        sdict[key]['TEMP']=sdict[key]["AHT_TEMP"]
    else:
        sdict[key]['TEMP']=(sdict[key]["AHT_TEMP"]+sdict[key]['DHT_TEMP'])/2
    
    print(key)
    print(sdict[key])
    print("_____________________________________________________________________________")
```

```python
for key,values in zip(sdict.keys(),sdict.values()):
    print(key)
    print(values.isnull().sum()/values.shape[0])
    
    print('_______________')
```
